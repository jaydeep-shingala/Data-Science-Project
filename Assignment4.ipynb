# Research question:

How have the win percentage of two major parties BJP(bjp) and Congress(cng) Changes over last years since bjp has been formed.

### Let's start visulization.

```python
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
bjp
```

```python
cng
```

#### Let's make one final dataframe from both: Merging data frame

```python
df = pd.DataFrame()
df['Year'] = bjp['Year']
```

```python
df
```

```python
df["Legislature"] = cng["Legislature"]
```

```python
df['Seats_won_bjp'] = bjp['Seats won']
df['Seats_won_cng'] = cng['Seats won']
df['Change_seat_bjp'] = bjp['Change in seats']
df['Change_seat_cng'] = cng['Change in seats']
df['per_vots_bjp'] = bjp['Percentage of votes']
df['per_vots_cng'] = cng['Percentage of votes']
df['vote_swing_bjp'] = bjp['Vote swing']
df['vote_swing_cng'] = cng['Vote swing']
df['outcome_bjp'] = bjp['Outcome']
df['outcome_cng'] = cng['Outcome']
```

```python
df
```

```python
plt.style.use('seaborn-darkgrid')
ax= plt.gca()
df.plot(kind='line',x='Year',y='Seats_won_bjp',ax=ax,label='BJP')
#plt.text(10.2, df.Seats_won_bjp.tail(1), 'BJP', horizontalalignment='left', size='small', color='b')
df.plot(kind='line',x='Year',y='Seats_won_cng', color='red',ax=ax,label='Congress')
plt.legend(frameon=False)
```
